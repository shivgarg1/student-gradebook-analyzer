# Student Gradebook Analyzer

Workshop project 2 - AIML using Python

This notebook analyzes a class gradebook using Pandas and NumPy. I have 10 students and their scores across 4 assignments. The plan is:
1. Load the data
2. Get stats for each student (avg/median/max/min)
3. Get stats for each assignment (avg/median/max/min)
4. Sort students to find top performers and who needs help
5. Use NumPy for standard deviation to see how spread out the scores are
6. Save everything to a new CSV

```python
import pandas as pd
import numpy as np
```

## 1. Loading the data

I made a small CSV file with 10 students and 4 assignments (out of 100 each). Honestly just typed in random-ish realistic looking scores for this, this isn't real data.

```python
df = pd.read_csv("gradebook.csv")
df
```

## 2. Per-student statistics

For each student, I need the average, median, max and min across their 4 assignments. The assignment columns are side by side (columns), so to get a stat *per row* (per student) I need `axis=1`. This confused me at first ngl, kept mixing up axis=0 and axis=1, but the way I remember it now is: axis=1 squishes the columns together for each row.

```python
assignment_cols = ["Assignment_1", "Assignment_2", "Assignment_3", "Assignment_4"]

df["Average"] = df[assignment_cols].mean(axis=1)
df["Median"] = df[assignment_cols].median(axis=1)
df["Max_Score"] = df[assignment_cols].max(axis=1)
df["Min_Score"] = df[assignment_cols].min(axis=1)

df
```

## 3. Per-assignment statistics

Now flip it - for each assignment, what's the average/median/max/min across all 10 students? This time it's column-wise so I don't even need to pass axis (default is axis=0).

```python
assignment_stats = df[assignment_cols].agg(["mean", "median", "max", "min"])
assignment_stats
```

Assignment_4 has the highest average (80.2) and Assignment_1 has the lowest (77.0). Makes sense if people got more comfortable as the assignments went on, or maybe Assignment_1 was just harder.

## 4. Sorting students by average score

```python
top_students = df.sort_values(by="Average", ascending=False)
top_students[["Student_Name", "Average"]]
```

### Top 3 performers

```python
top_3 = top_students.head(3)
top_3[["Student_Name", "Average"]]
```

### Bottom 3 (might need extra support)

Same thing, sorted the other way. Could be useful if a teacher wanted to know who to check in with.

```python
bottom_3 = df.sort_values(by="Average", ascending=True).head(3)
bottom_3[["Student_Name", "Average"]]
```

## 5. Using NumPy properly

I imported numpy earlier but wasn't actually using it for anything (pandas was doing all the work). So here I'm pulling the scores out into an actual NumPy array with `.to_numpy()` and calculating standard deviation - this tells us how spread out the scores are. A high std means some students scored really differently from others on that assignment, a low std means everyone scored pretty similarly.

```python
scores_array = df[assignment_cols].to_numpy()

class_average_np = np.mean(scores_array)
class_std_np = np.std(scores_array)

print(f"Overall class average (via NumPy): {class_average_np:.2f}")
print(f"Overall class standard deviation (via NumPy): {class_std_np:.2f}")
```

```python
# std per assignment, axis=0 because I want one number per column this time
assignment_std = np.std(scores_array, axis=0)

for col_name, std_val in zip(assignment_cols, assignment_std):
    print(f"{col_name}: {std_val:.2f}")
```

## 6. Saving the results

```python
df.to_csv("gradebook_with_stats.csv", index=False)
print("Saved to gradebook_with_stats.csv")
```
